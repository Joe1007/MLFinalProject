```python
def process_wafer_map(df):
    # 創建空的 list 來存儲處理後的資料
    processed_data = []
    
    # 處理每一個 wafer map
    for wafer_map in df['waferMap']:
        # 將 wafer map 轉換為 numpy array 並攤平
        flattened = np.array(wafer_map).flatten()
        processed_data.append(flattened)
    
    # 創建新的 DataFrame，包含 V1 到 V676
    wafer_df = pd.DataFrame(processed_data, columns=[f'V{i+1}' for i in range(676)],
                           index=df.index)  # 添加這行，確保索引一致
    
    # 加入原本的 trianTestLabel 和 failureType
    result_df = pd.concat([wafer_df, df[['trianTestLabel', 'failureType']]], axis=1)
    
    return result_df

# 使用函數處理整個 dataframe
result_df = process_wafer_map(df4)
result_df

```

## SMOTE

```python
import pandas as pd
from collections import Counter

# 載入資料集
data = pd.read_csv('WM811K.csv')

# 查看資料的前幾列和類別分布
print(data.head())
print(data['failureType'].value_counts())
```

```python
# 移除樣本數少於 5 的類別
filtered_data = data[data['failureType'].map(data['failureType'].value_counts()) >= 5]
X = filtered_data.drop(columns=['failureType', 'Unnamed: 0', 'trianTestLabel'])
y = filtered_data['failureType']

```

```python
from imblearn.combine import SMOTEENN
from collections import Counter

# 使用 SMOTEENN 進行混合採樣
smote_enn = SMOTEENN(random_state=42)
X_resampled, y_resampled = smote_enn.fit_resample(X, y)

# 檢查平衡後的類別分布
print("平衡後的類別分布：", Counter(y_resampled))
```

```python
balanced_data = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.DataFrame(y_resampled, columns=['failureType'])], axis=1)
balanced_data
balanced_data.to_csv('WM811K_balanced.csv', index=True)

```
